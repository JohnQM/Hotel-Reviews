```python
tokenized_reviews = []
for item in hrd["reviews.text"]:
    tokenized_reviews.append(item.split())
```

```python
print(tokenized_reviews[1:10])
```

```python
# after breaking up the reviews or sentences into individual words, the tokens require additional processing to remove excess punctuation
punctuation = [",", ":", ";", ".", "'", '"', "’", "?", "/", "-", "+", "&", "(", ")"]
new_tokenized = []
for item in tokenized_reviews:
    tokens = []
    for token in item:
        token = token.lower()
        for punc in punctuation:
            token = token.replace(punc, "")
        tokens.append(token)
    new_tokenized.append(tokens)
```

```python
# after lowercasing and remove the punctuation, the new dataset has gone through preprocessing in order for a more accurate prediction
print(new_tokenized[1:10])
```

```python
# the next step involves capturing all of the unique words within the hotel reviews or any tokens that occur more than once across all of the reviews
unique_tokens = []
single_tokens = []
for tokens in new_tokenized:
    for token in tokens:
        if token not in single_tokens:
            single_tokens.append(token)
        elif token in single_tokens and token not in unique_tokens:
            unique_tokens.append(token)

counts = pd.DataFrame(0, index=np.arange(len(new_tokenized)), columns=unique_tokens)
```

```python
for i, item in enumerate(new_tokenized):
    for token in item:
        if token in unique_tokens:
            counts.iloc[i][token] += 1
```

```python
word_counts = counts.sum(axis=0)
counts = counts.loc[:,(word_counts >= 5) & (word_counts <= 100)]
```

```python
# Now we'll need to split the data into two sets so that we can evaluate our algorithm effectively. We'll train our algorithm on a training set, then test its performance on a test set
# In this mission, we'll be predicting the number of ratings the hotels received, based on their reviews. 
# Because ratings are an indicator of popularity, we'll discover which types of reviews tend to be the most popular.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(counts, hrd["reviews.rating"], test_size=0.3, random_state=1)
```

```python
# Essentially, the model is determining which words correlate with more ratings and which with less. 
# By finding these correlations, the model will be able to predict which reviews will receive higher ratings in the future. 
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
```

```python
mse = sum((predictions - y_test) ** 2) / len(predictions)
```
